```python
df = pd.read_csv('data/kr_korean.csv', names=['단어', '품사'])

df.head()
```

```python
df = df[df['품사'] == '명사']

df.head()
```

```python
df = df.drop_duplicates(subset=['단어'])

df.head()
```

```python
df.drop(df['단어'].filter(regex='/[^가-힣]/g'), axis=1)
```

```python
df = df[
    (~df['단어'].str.contains('[^가-힣]')) &
    (df['단어'].str.len() >= 2)
]

df.head()
```

```python
df.to_csv('data/kr_korean_nouns.csv', index=None, encoding='utf-8')
```
